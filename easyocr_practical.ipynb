```python
import easyocr
```

```python
reader = easyocr.Reader(['en'])
```

```python
result = reader.readtext('/content/OCR test.png')
```

```python
for (bbox, text, prob) in result:
    print(f'Text: {text}, Probability: {prob}')
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/OCR test.png')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
for (bbox, text, prob) in result:
    print(f"Detected text: {text} (Confidence: {prob:.2f})")
```

```python
for (bbox, text, prob) in result:
    # Get the bounding box coordinates
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple([int(val) for val in top_left])
    bottom_right = tuple([int(val) for val in bottom_right])

    # Draw the bounding box
    cv2.rectangle(image_rgb, top_left, bottom_right, (0, 255, 0), 2)

    # Put the detected text
    cv2.putText(image_rgb, text, (top_left[0], top_left[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

# Display the annotated image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
